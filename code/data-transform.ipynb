```python
import pandas as pd
import re
from pathlib import Path

def parse_duration(d):
    match = re.match(r'PT(?:(\d+)H)?(?:(\d+)M)?(?:(\d+)S)?', d)
    hours = int(match.group(1) or 0)
    minutes = int(match.group(2) or 0)
    seconds = int(match.group(3) or 0)
    if hours > 0:
        return f"{hours:02}:{minutes:02}:{seconds:02}"
    else:
        return f"{minutes:02}:{seconds:02}"

df=pd.read_json('../datastore/videos.json')
display(df)
df['publishedAt']=pd.to_datetime(df['publishedAt']).dt.tz_localize(None)
df['duration'] = df['duration'].apply(parse_duration)
display(df,df.info())

if not Path('../datastore/videos_list.xlsx').exists():
    df.to_excel('../datastore/videos_list.xlsx',index_label='id')
```
